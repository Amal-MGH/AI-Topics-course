```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
#1. Load the dataset from CSV file.
dataset=pd.read_csv('C:/Users/amal/Desktop/diabetes_csv.csv')#using panda with the dataset path

#2. Divide the dataset into training and testing.

X = dataset.iloc[:, :-1].values #X contaend all row with all featurs
y = dataset.iloc[:, 8].values   #y contaend all row with the class only


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30) #sperate the X (features) and y(class)
                                                                          #to 70% training & 30% testing
                                                                          #using train_test_split


hAccuracy=-1  #varabile to save the higher accuracy
K=0           #varabile to save the K value that give higher accuracy


print('\n********************************* KNN Classifiers*********************************\n\n')



for k in [1,3,5,10]:    # loop will go over all possiable value for k to test it
    
    #3. Train the KNN algorithm.
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X_train, y_train)
    
    #4. Test the algorithm on the test dataset with different values of k=1,3,5,10.
    y_pred = classifier.predict(X_test)
    
    #5. Report the accuracy, precision, and recall.
    from sklearn.metrics import accuracy_score,precision_score,recall_score
    
    accuracy=accuracy_score(y_test, y_pred)                       #accuracy=num.correct/num.testSet
    precision= precision_score(y_test,y_pred,average='macro')     #precision=num.correct/num.testSet
    recall=recall_score(y_test,y_pred,average='macro')            #recall=num.correct/num.testSet
    
    print('when k={}'.format(k))
    print ('Accuracy  : {:.4f}%'.format(accuracy*100))
    print ('Precision : {:.4f}%'.format(precision*100))
    print ('Recall    : {:.4f}%'.format(recall*100))
    print('_____________________\n')
    
    if (accuracy>hAccuracy):   #check accuracy to save the higher one with its k
        hAccuracy=accuracy
        K=k
        

#6. Report which k that gives the highest accuracy
print('\nthe k value that give highest accuracy is : {}, accuracy :{:.4f}%\n'.format(K,hAccuracy*100))

```
